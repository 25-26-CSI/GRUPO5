#EJEMPLOS DE TECNICAS DE CIFRADO

##CIFRADO POR TRANSPOSICI√ìN

```python
#CIFRADO POR TRANSPOSICI√ìN

def cifrar_transposicion(mensaje, clave):
    # Eliminar espacios
    mensaje = mensaje.replace(" ", "").upper()

    # Determinar n√∫mero de columnas
    num_columnas = len(clave)

    # Determinar n√∫mero de filas (redondeando hacia arriba)
    num_filas = -(-len(mensaje) // num_columnas)  # divisi√≥n techo

    # Rellenar con "X" si falta completar la √∫ltima fila
    while len(mensaje) < num_columnas * num_filas:
        mensaje += "X"

    # Crear la matriz del mensaje
    matriz = [list(mensaje[i:i+num_columnas]) for i in range(0, len(mensaje), num_columnas)]

    # Reordenar las columnas seg√∫n la clave
    columnas_ordenadas = sorted(list(enumerate(clave)), key=lambda x: x[1])

    # Crear el texto cifrado leyendo las columnas en el orden indicado por la clave
    texto_cifrado = ""
    for idx, _ in columnas_ordenadas:
        for fila in matriz:
            texto_cifrado += fila[idx]

    return texto_cifrado, matriz


def descifrar_transposicion(texto_cifrado, clave):
    num_columnas = len(clave)
    num_filas = -(-len(texto_cifrado) // num_columnas)

    # Calcular el n√∫mero de letras por columna
    letras_por_columna = num_filas

    # Reordenar las columnas seg√∫n la clave
    columnas_ordenadas = sorted(list(enumerate(clave)), key=lambda x: x[1])

    # Crear lista vac√≠a para las columnas
    columnas = [''] * num_columnas

    # Rellenar cada columna con sus letras
    pos = 0
    for idx, _ in columnas_ordenadas:
        columnas[idx] = texto_cifrado[pos:pos+letras_por_columna]
        pos += letras_por_columna

    # Reconstruir el mensaje fila por fila
    texto_descifrado = ''
    for i in range(num_filas):
        for j in range(num_columnas):
            texto_descifrado += columnas[j][i]

    return texto_descifrado


# INTERACCI√ìN CON EJEMPLO

mensaje = input("Texto a cifrar: ").strip()
clave_str = input("Ingresa la clave num√©rica separada por espacios (ejemplo: 3 1 4 2): ")

# Convertir la clave a lista de n√∫meros
clave = [int(x) for x in clave_str.split()]

# Cifrar el mensaje
cifrado, matriz = cifrar_transposicion(mensaje, clave)

print("\n MATRIZ USADA PARA EL CIFRADO:")
for fila in matriz:
    print(fila)

print("\nüîí Texto cifrado:", cifrado)

# Descifrar el mensaje
descifrado = descifrar_transposicion(cifrado, clave)
print("üîì Texto descifrado:", descifrado)
```
