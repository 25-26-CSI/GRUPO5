#EJEMPLOS DE TECNICAS DE CIFRADO

##CIFRADO POR SUSTITUCIÓN

```python
def cifrar_sustitucion(mensaje, desplazamiento):
    alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    mensaje = mensaje.upper()
    cifrado = ""

    for letra in mensaje:
        if letra in alfabeto:
            indice = (alfabeto.index(letra) + desplazamiento) % 26
            cifrado += alfabeto[indice]
        else:
            cifrado += letra  # mantiene espacios, comas, etc.
    return cifrado


def descifrar_sustitucion(mensaje_cifrado, desplazamiento):
    alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    mensaje_cifrado = mensaje_cifrado.upper()
    descifrado = ""

    for letra in mensaje_cifrado:
        if letra in alfabeto:
            indice = (alfabeto.index(letra) - desplazamiento) % 26
            descifrado += alfabeto[indice]
        else:
            descifrado += letra
    return descifrado


# Ejemplo de uso
texto = input("Escribe el mensaje a cifrar: ")
clave = int(input("Escribe el número de desplazamiento (clave): "))

mensaje_cifrado = cifrar_sustitucion(texto, clave)
print("Mensaje cifrado:", mensaje_cifrado)

mensaje_descifrado = descifrar_sustitucion(mensaje_cifrado, clave)
print("Mensaje descifrado:", mensaje_descifrado)
```
