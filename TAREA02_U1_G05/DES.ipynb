```python
from Crypto.Cipher import DES
from Crypto.Random import get_random_bytes
import base64

# Generar una clave de 8 bytes (DES usa claves de 64 bits)
key = get_random_bytes(8)

# Crear objeto de cifrado DES (modo ECB por simplicidad)
cipher = DES.new(key, DES.MODE_ECB)

# Mensaje a cifrar (debe tener longitud múltiplo de 8)
text = "hola uce"
data = text.encode('utf-8')

# Cifrar el texto
encrypted_data = cipher.encrypt(data)
print("Texto cifrado (base64):", base64.b64encode(encrypted_data).decode())

# Descifrar
decipher = DES.new(key, DES.MODE_ECB)
decrypted_data = decipher.decrypt(encrypted_data)
print("Texto descifrado:", decrypted_data.decode('utf-8'))
```
